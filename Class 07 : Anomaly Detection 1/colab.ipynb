```python
def IOR_Detection(df):
    for col in df.columns:
        Q3 = df[col].quantile(0.25)
        Q1 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        min_value = Q3 - IQR*1.5
        max_value = Q1 + IQR*1.5
   
```

```python
def z_score_outlier(df,threshold=2):
    outliers = {}
    for col in df.columns:
       z = (df[col] - df[col].mean())/df[col].std()
       outliers[col] =  df[abs(z) > threshold].index.tolist()
    return outliers


```
