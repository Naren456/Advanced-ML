```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
```

```python

featue1 = np.random.rand(100)
featue2 = np.random.rand(100)*1000   # Weight (40-120 kg)
featue3 = np.random.rand(100) * 50   # Height (150-200 cm)
```

```python
df = pd.DataFrame({'Feature1':featue1,'Feature2':featue2,'Feature3':featue3})
```

```python
from sklearn.decomposition import PCA

pca_no_scale = PCA()
pca_no_scale.fit(df)

print(np.round(pca_no_scale.explained_variance_ratio_,3))
```

```python
from sklearn.preprocessing import StandardScaler

Scaler = StandardScaler()
scaled_data = Scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

pca_no_scale = PCA()
pca_no_scale.fit(scaled_data)

print(np.round(pca_no_scale.explained_variance_ratio_,3))
```
